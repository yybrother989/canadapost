# Model Implementing

## Prophet

```python
processed_df = df[(df.City == 'City1')& (df.Customer.isin(target_customer))]
```

```python
processed_df = processed_df[['Date','Volume']]
```

```python
processed_df = processed_df.rename({'Date':'ds','Volume':'y'},axis = 1)
```

```python
processed_df = processed_df[processed_df['ds']< '2022-09-01']
```

```python
period = df['Date'].max()-dt.datetime(2022,9,1)
```

```python
processed_df.sort_values('ds')
```

```python
ph = Prophet()
```

```python
ph.fit(processed_df)
```

```python
future = ph.make_future_dataframe(periods=period.days)
```

```python
future.tail()
```

```python
forecast = ph.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
pred = forecast[forecast.ds>='2022-09-01']
```

```python
pred
```

```python
pred['DAY_NAME'] = pd.DatetimeIndex(pred.loc[:,'ds']).day_name()
```

```python
test_df = df[(df.City == 'City1')& (df.Customer.isin(target_customer))&(df.Date>'2022-09-01')]
```

```python
plt.figure(figsize=(20,10))
test_df.groupby('DAY_NAME').sum('Volume')['Volume'].plot(label  = 'true')
pred.groupby('DAY_NAME')['yhat'].sum().plot(label = 'predict')

plt.legend()
plt.show()
```

```python
plt.figure(figsize=(20,10))
sns.lineplot(x = 'Date', y = 'Volume', data = test_df,ci=False,label = 'True')
sns.lineplot(x = 'ds', y = 'yhat', data = pred,label = 'Predicted')
plt.xticks(rotation = 90)
plt.show()
```
